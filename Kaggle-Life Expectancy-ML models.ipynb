```python
def missing_values(df):
    missing = df.isnull().sum().sort_values(ascending=False)
    missing = missing[missing > 0]
    missing_percent = (missing/len(df))*100
    return pd.DataFrame({'Missing Values': missing, 'Missing Percentage': missing_percent})
```

```python
def separate_features(df):
    categorical_features = df.select_dtypes(include='object').columns
    numerical_features = df.select_dtypes(exclude='object').columns
    return [categorical_features, numerical_features]
```

```python
X = df.drop('Life expectancy ',axis=1)
y = df['Life expectancy ']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
scaler = StandardScaler()
scaled_xtrain = scaler.fit_transform(X_train)
scaled_xtest = scaler.transform(X_test)
```

```python
# Creating a function that will take in the model, fits, predicts and returns the error metrics
def evaluate_regression_model(model):
    model.fit(scaled_xtrain,y_train)
    
    y_train_pred = model.predict(scaled_xtrain)
    y_test_pred = model.predict(scaled_xtest)
    
    y_mean = np.mean(y)
    
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_rmse = np.sqrt(train_mse)
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_mae_rel = train_mae / y_mean *100
    
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_rmse = np.sqrt(test_mse)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_mae_rel = test_mae / y_mean*100
    
    return [train_mse, train_rmse, train_mae, train_mae_rel], [test_mse, test_rmse, test_mae, test_mae_rel]
```

```python
dec_treemod = DecisionTreeRegressor()
rfr_mod = RandomForestRegressor()
gbr_mod = GradientBoostingRegressor()
xgb_mod = XGBRegressor()
cat_mod = CatBoostRegressor()
linear_mod = LinearRegression()
```

```python
evaluate_regression_model(linear_mod)
```

```python
evaluate_regression_model(rfr_mod)
```

```python
evaluate_regression_model(xgb_mod)
```

```python
evaluate_regression_model(cat_mod)
```

```python
evaluate_regression_model(dec_treemod)
```

```python
# Random Forest Model performs the best compared to all of them

Random_forest_model = RandomForestRegressor(n_jobs=-1)
param_grid = {'n_estimators':[10,50,100,200,500],
             'max_depth':[None,2,4,6,8],
             'max_features':['auto','sqrt'],
             'min_impurity_decrease':[0.0,0.2,0.5],
             'max_leaf_nodes':[None,2,]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
Grid_model = GridSearchCV(Random_forest_model,param_grid,cv=5,n_jobs=-1,verbose=2)
```

```python
Grid_model.fit(scaled_xtrain,y_train)
```

```python
Grid_model.best_estimator_
```

```python
Grid_model.best_params_
```

```python
train_preds = Grid_model.predict(scaled_xtrain)
train_mae = mean_absolute_error(y_train,train_preds)
train_mse = mean_squared_error(y_train,train_preds)
train_rmse = np.sqrt(train_mae)

test_preds = Grid_model.predict(scaled_xtest)
test_mae = mean_absolute_error(y_test,test_preds)
test_mse = mean_squared_error(y_test,test_preds)
test_rmse = np.sqrt(test_mae)

print('TRAIN MAE',train_mae), 
print('TRAIN MSE',train_mse), 
print('TRAIN RMSE',train_rmse) 
print('\n')
print('TEST MAE',test_mae),
print('TEST_MSE',test_mse),
print('TEST RMSE',test_rmse)
```
